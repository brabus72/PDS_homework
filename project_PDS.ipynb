```python
def correct_rooms(df):
    df.loc[(df['Rooms'] == 0) & (df['Square'] <= 50), 'Rooms'] = 1
    df.loc[(df['Rooms'] == 0) & ((df['Square'] > 50) & (df['Square'] <= 100)), 'Rooms'] = 2
    df.loc[(df['Rooms'] == 0) & ((df['Square'] > 100) & (df['Square'] <= 150)), 'Rooms'] = 3
    df.loc[(df['Rooms'] == 0) & ((df['Square'] > 150) & (df['Square'] <= 250)), 'Rooms'] = 4
    df.loc[(df['Rooms'] == 0) & ((df['Square'] > 250) & (df['Square'] <= 400)), 'Rooms'] = 5
    df.loc[(df['Rooms'] == 0) & (df['Square'] > 400), 'Rooms'] = 6
    df.loc[(df['Rooms'] >= 6) & (df['Square'] <= 50), 'Rooms'] = 1
    df.loc[(df['Rooms'] >= 6) & ((df['Square'] > 50) & (df['Square'] <= 100)), 'Rooms'] = 2
    df.loc[(df['Rooms'] >= 6) & ((df['Square'] > 100) & (df['Square'] <= 150)), 'Rooms'] = 3
    df.loc[(df['Rooms'] >= 6) & ((df['Square'] > 150) & (df['Square'] <= 250)), 'Rooms'] = 4
    df.loc[(df['Rooms'] >= 6) & ((df['Square'] > 250) & (df['Square'] <= 400)), 'Rooms'] = 5
    df.loc[(df['Rooms'] >= 6) & (df['Square'] > 400), 'Rooms'] = 6
    df = add_room_size(df, room_size)
    return df
```
